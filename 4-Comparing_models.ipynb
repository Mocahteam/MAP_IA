# Comparaison des résultats des différentes approches

Ce notebook compare les résultats des approches :
- Transformer avec attention (voir `2-MAP_IA_Seq2Seq_TransformerWithAttention.ipynb`)
- LSTM (voir `2-MAP_IA_Seq2Seq_LSTM.ipynb`)
- MAP approche algorithmique (voir `MAP_Analysis.ipynb` dans le répository de MAP)

## Chargement des résultats

```python
import json

file = open('exportResults_Seq2Seq_TransformerWithAttention.txt', 'r')
Seq2Seq_TWA = json.loads(file.readline())
file.close()
file = open('exportResults_Seq2Seq_LSTM.txt', 'r')
Seq2Seq_LSTM = json.loads(file.readline())
file.close()
file = open('exportResults_MAP_hard.txt', 'r')
MAP_hard = json.loads(file.readline())
file.close()
file = open('exportResults_MAP_soft.txt', 'r')
MAP_soft = json.loads(file.readline())
file.close()

file = open('exportResults_alter_Seq2Seq_LSTM.txt', 'r')
Seq2Seq_LSTM_alter = json.loads(file.readline())
file.close()
file = open('exportResults_MAP_alter_hard.txt', 'r')
MAP_hard_alter = json.loads(file.readline())
file.close()
file = open('exportResults_MAP_alter_soft.txt', 'r')
MAP_soft_alter = json.loads(file.readline())

file = open('exportResults_v2_Seq2Seq_LSTM.txt', 'r')
Seq2Seq_LSTM_v2 = json.loads(file.readline())
file.close()
file = open('exportResults_MAP_v2_hard.txt', 'r')
MAP_hard_v2 = json.loads(file.readline())
file.close()
file = open('exportResults_MAP_v2_soft.txt', 'r')
MAP_soft_v2 = json.loads(file.readline())
file.close()
```

## Affichage des résultats

```python
import matplotlib.pyplot as plt
import numpy as np

# Position des barres sur l'axe x
x = np.arange(len(MAP_hard_v2))
largeur = 0.25  # largeur des barres

# Création des barres
fig, ax = plt.subplots(figsize=(12, 3))
bar1 = ax.bar(x - largeur, MAP_hard_v2, largeur, label='MAP algorithm (hard)')
bar2 = ax.bar(x, MAP_soft_v2, largeur, label='MAP algorithm (soft)')
bar3 = ax.bar(x + largeur, Seq2Seq_LSTM_v2, largeur, label='Seq2Seq_LSTM')
#bar3 = ax.bar(x + largeur, Seq2Seq_TWA, largeur, label='Seq2Seq_TWA')

# Configuration du graphique
ax.set_xlabel('Scenarios')
ax.set_ylabel('Number of validated tests')
#ax.set_title('Histogramme de la performance des approches')
ax.set_xticks(x)
# Pour la version alter
# ax.set_xticklabels(["1-[A]", "2-A[B]C", "3-A[BC]D", "4-A[BCD]E", "5-[A][B]", "6-[A[B]]", "7-[[A]B]", "8-[A[BC]]", "9-[[AB]C]", "10-A[B]C...D[E]F", "11-A[B][C[D]]", "12-[AB*C]", "13-[A*BCA*]", "14-[[AB]*C[AB]*]", "15-[A[BC*D]]", "16-[A[B*CB*]"], rotation=45, ha='right')
# Pour la version v2
ax.set_xticklabels(["S0", "S1", "S2", "S3", "S4", "S5", "S6", "S7", "S8", "S9"], ha='right')
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Position des barres sur l'axe x
x = np.arange(len(MAP_hard_alter))
largeur = 0.25  # largeur des barres

# Création des barres
fig, ax = plt.subplots(figsize=(12, 6))
bar1 = ax.bar(x - largeur, MAP_hard_alter, largeur, label='MAP algorithm (hard)')
bar2 = ax.bar(x, MAP_soft_alter, largeur, label='MAP algorithm (soft)')
bar3 = ax.bar(x + largeur, Seq2Seq_LSTM_alter, largeur, label='Seq2Seq_LSTM')
#bar3 = ax.bar(x + largeur, Seq2Seq_TWA, largeur, label='Seq2Seq_TWA')

# Configuration du graphique
ax.set_xlabel('Scénarios')
ax.set_ylabel('Nombre de tests passés')
ax.set_title('Histogramme de la performance des approches')
ax.set_xticks(x)
ax.set_xticklabels(["1-[AB+]", "2-AB+[C]DE+", "3-[AB*C*D]", "4-[A*BA*]", "5-[A][B][C]", "6-[AB][C]"], rotation=45, ha='right')
ax.legend()

plt.show()
```
